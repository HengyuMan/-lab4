```python
import keras
from keras.layers import Input, Dense
from keras.models import Model
from keras import regularizers
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
from math import ceil, sqrt

```

```python
bin_train = np.genfromtxt('/Users/manhengyu/Desktop/ANN/project4/bindigit_trn.csv',delimiter=',',)
target_train = np.genfromtxt('/Users/manhengyu/Desktop/ANN/project4/targetdigit_trn.csv',delimiter=',',)
bin_test = np.genfromtxt('/Users/manhengyu/Desktop/ANN/project4/bindigit_tst.csv',delimiter=',',)
target_test = np.genfromtxt('/Users/manhengyu/Desktop/ANN/project4/targetdigit_tst.csv',delimiter=',',)
```

```python
new_data = np.reshape(bin_train[1,:],(28,28))
plt.imshow(new_data)
plt.gray()
```

# node=100

```python
epochstime = 50
hidden_dim = 100
batch_size = 64
input_img = Input(shape=(784,))
encoded = Dense(hidden_dim, activation='relu',kernel_initializer='random_uniform',bias_initializer='zeros')(input_img)
decoded = Dense(784, activation='hard_sigmoid',kernel_initializer='random_uniform',bias_initializer='zeros')(encoded)

autoencoder = Model(input_img, decoded)

autoencoder.compile(optimizer='adadelta', loss='mean_squared_error')

result = autoencoder.fit(bin_train, bin_train,epochs=epochstime,batch_size=batch_size,
                     shuffle=True,
                     verbose=1,
                     validation_data=(bin_test, bin_test))
MSEmat=result.history['loss']
valmse=result.history['val_loss']
    
```

```python
l1, = plt.plot(MSEmat, label='test set')
l2, = plt.plot(valmse, color='red', linewidth=1.0, linestyle='--', label='validation set')
plt.legend(loc='upper right')
plt.xlabel('epoch times')
plt.ylabel('MSE')
plt.title('MSE-epoch curve')
plt.show
```

```python
prediction = autoencoder.predict(bin_test)
plt.figure(figsize = (20,10))
for j in range(1,10):
    index = np.where(target_test==j)[0][0]
    # Plot original
    plt.subplot(10, 2, j*2-1)
    plt.imshow(bin_test[index,:].reshape((28, 28)))
    plt.gray()
    plt.xticks(())
    plt.yticks(())

    plt.subplot(10, 2, j*2)
    #Plot reconstruction
    plt.imshow(prediction[index,:].reshape((28, 28)))
    plt.gray()
    plt.xticks(())
    plt.yticks(())
plt.suptitle('Comparision between original and prediction')
plt.subplots_adjust(left = 0.6, right = 0.7, wspace = 0.5, hspace = 0.5)
#Plot Autoencoder weights
weights = autoencoder.get_weights()[0].T
size = weights.shape[0]
plt.figure(figsize=(10.2, 10))
for i in range(size):
    ax = plt.subplot(10, ceil(size/10), i + 1)
    plt.imshow(weights[i].reshape(28, 28))
    plt.gray()
    plt.xticks(())
    plt.yticks(()) 
plt.suptitle('{} components extracted by autoencoder'.format(hidden_dim))
```

```python
def sparse(x):
    a=sum(np.abs(x))
    n=len(x)
    b=sqrt(sum(np.power(x,2)))
    return (sqrt(n)-a/b)/(sqrt(n)-1)    
dense1_layer_model = Model(inputs=autoencoder.input,outputs=autoencoder.get_layer('Dense_1').output)
encodeoutput=dense1_layer_model.predict(bin_train)
sparseness=[]
for i in range(encodeoutput.shape[0]):
    sparseness.append(sparse(encodeoutput[i]))
print(np.mean(sparseness))
                      


```
